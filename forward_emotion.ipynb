```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.pyplot import plot, scatter
```

```python
g = 9.81
x0 = 0
v0 = 2.6
```

## Settings

```python
t_max = 0.5 # eigentlich 5.5
n =  100
dt = t_max/n

t = np.linspace(0, t_max, n+1)
index_list = np.linspace(0, n-1, n, dtype=int)
```

## Solving

```python
v = [v0]
vk = v0

for k in index_list:
    vk_plus_1 = -dt*g + vk
    v.append(vk_plus_1)
    vk = vk_plus_1

v = np.array(v)
```

## Postprocessing

```python
scatter(t, v)
plt.grid(True)
```

### Vorzeichenwechsel der Geschwindigkeit

```python
index_min = np.argmin(abs(np.array(v)))
t[index_min]
```

## Weg

```python
x = [x0]
xk = x0
for k, time in enumerate(np.linspace(0, n-1, n, dtype=int)):
    xk_plus_1 = dt*v[k] + xk
    x.append(xk_plus_1)
    xk = xk_plus_1
```

```python
scatter(t, x)
```
